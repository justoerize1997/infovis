```python
import pandas as pd
import altair as alt

df = pd.read_csv("https://raw.githubusercontent.com/aaizemberg/ecd2022a/main/s2/2022w44.csv")

alt.Chart(df).mark_circle(size=60,opacity=0.5).encode(
    x='raised:Q',
    y='spent:Q',
    color='party:N',
    tooltip=['member', 'chamber', 'party', 'raised', 'spent']
).interactive()
```

```python
alt.Chart(df).mark_tick().encode(
    x='raised:Q',
    y='party:N',
    color='party',
    tooltip=['member', 'chamber', 'party', 'raised', 'spent']
).interactive()
```
